```python
import pandas as pd
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
import pickle as pk
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('C:/Users/user/ashish/ml/Model_big_mart/big_mart.csv')
```

## Data Preprocessing

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Item_Weight']=df['Item_Weight'].fillna(df['Item_Weight'].mean())
```

```python
df['Outlet_Size']=df['Outlet_Size'].fillna(list(df['Outlet_Size'].mode())[0])
```

```python
df['Item_Fat_Content'].value_counts()
```

```python
df['Item_Fat_Content']=df['Item_Fat_Content'].replace({'LF':'Low Fat','low fat':'Low Fat','reg':'Regular'})
```

## Model Training

```python
df.head()
```

```python
df.tail()
```

```python
encoder=LabelEncoder()
```

```python
df['Item_Identifier'] = encoder.fit_transform(df['Item_Identifier'])

df['Item_Fat_Content'] = encoder.fit_transform(df['Item_Fat_Content'])

df['Item_Type'] = encoder.fit_transform(df['Item_Type'])

df['Outlet_Identifier'] = encoder.fit_transform(df['Outlet_Identifier'])

df['Outlet_Size'] = encoder.fit_transform(df['Outlet_Size'])

df['Outlet_Location_Type'] = encoder.fit_transform(df['Outlet_Location_Type'])

df['Outlet_Type'] = encoder.fit_transform(df['Outlet_Type'])
```

```python
X=df.drop('Item_Outlet_Sales',axis=1)      #Features
Y=df['Item_Outlet_Sales']                   #Target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.18,random_state=1)
```

```python
rgr=XGBRegressor(enable_categorical=True)
```

```python
rgr.fit(X_train,Y_train)
```

```python
pred_train=rgr.predict(X_train)
pred_test=rgr.predict(X_test)
```

```python
diff_train=r2_score(Y_train,pred_train)
```

```python
diff_train
```

```python
diff_test=r2_score(Y_test,pred_test)
```

```python
diff_test
```
